# Data Preprocessing

```python
df.info()
```

```python
id_ = 'ID'
default = 'default.payment.next.month'
categorical = ['SEX', 'EDUCATION', 'MARRIAGE']
numerical = [col for col in df.columns if df[col].dtype == "int64" or df[col].dtype == "float64"]
numerical.remove('ID')
numerical.remove('default.payment.next.month')
numerical
```

```python
# One Hot Encoder
df_cat = pd.get_dummies(df[categorical], drop_first = True)
df_cat.head()
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_num = scaler.fit_transform(df[numerical])
df_num = pd.DataFrame(df_num, columns=numerical)
df_num.describe()
```

```python
df_preprocessed = pd.concat([df.ID, df_num, df_cat, df[default]], axis=1)
df_preprocessed.head()
```

```python
pay = ['PAY_0'] + [f'PAY_{i}' for i in range(2, 7)]
df_preprocessed = df_preprocessed.drop(pay,axis=1)
```

```python
df_preprocessed.head()
```

# Ini Untuk Pemodelan

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X = df_preprocessed.drop(default, axis = 1).values
y = df_preprocessed[default].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
dt.get_params()
```

```python
param_grid = {
    'max_depth': [i for i in range(3, 10, 2)]
}
dt = DecisionTreeClassifier(random_state=42)
clf = GridSearchCV(dt, param_grid, cv=5)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
clf.score(X_test, y_test)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
plot_tree(clf.best_estimator_, fontsize=4)
plt.show()
```
